```python
# General Imports:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab
from scipy.optimize import fmin, minimize 
import scipy.optimize
from math import exp, inf
from itertools import chain
```

```python
messages = ["GOOG_2012-06-21_34200000_57600000_message_5.csv"]
order_book = ["GOOG_2012-06-21_34200000_57600000_orderbook_5.csv"]
mess = pd.read_csv(messages[0])
book = pd.read_csv(order_book[0])
len(book.columns)
```

```python
mess.columns = ["ts", "type", "id", "size_price", "price", "direction"]
a = ["ask_price_1", "ask_size_1", 'bid_price_1', 'bid_size_1',
     "ask_price_2", "ask_size_2", 'bid_price_2', 'bid_size_2',
     "ask_price_3", "ask_size_3", 'bid_price_3', 'bid_size_3',
     "ask_price_4", "ask_size_4", 'bid_price_4', 'bid_size_4',
     "ask_price_5", "ask_size_5", 'bid_price_5', 'bid_size_5'   
]
book.columns = a
```

```python
book.head()
```

```python
mess["index"] = mess.index
book["index"] = book.index
df_trades = pd.merge(mess, book)
df_trades["mid_price"] = (df_trades.ask_price_1 + df_trades.bid_price_1)/2
del df_trades['id']
del df_trades['index']
#supported_types = [1, 2, 3, 4]
df_trades = df_trades[df_trades.type <= 4].reset_index(drop=True)
# Les ordres invisibles ne peuvent pas être pris en compte
```

```python
df_trades.head(10)
# We notice that the data frame has already been filtered to trading hours
```

```python
def get_tick(df_trades_source):
    diffs = abs(df_trades_source.ask_price_1.diff(1).dropna())
    diffs = diffs[diffs != 0].round(3)
    return min(diffs)
tick = get_tick(df_trades)
# Dans notre cas, le tick est de l'ordre de 100.
global_AES = df_trades.size_price.mean()
print(global_AES)
```

# Modèles  Utilisés

## Modèle 1 : Modèle à queues indépendantes.

Comme on va directement le remarquer ici, on travailler avec des ticks qui sont très petits: le nombre de fois où la différence entre le best ask et le best bid est de un tick est quasiment de  178/112635), même le cas où l'un des deux best bid best asks est non nulle est très rare. Ce que nous allons faire c'est considérer que les deux best bids best asks sont toujours de quantités non nulles. On travaillera avec les limites non pas en terme de $p_{mid} + k  \delta$ avec $\delta$ le tick, mais plutôt avec les limites données par le fournisseur de données.

De plus, le modèle 1 suppose une symétrie. Nous allons donc utiliser indéfiniment les bids et les asks de chaque limite de la même manière.

```python
diff = df_trades.ask_price_1 - df_trades.bid_price_1
print(len(diff[diff == min(diff)]))
print(len(diff[diff <= min(diff) + 2*tick]))
```

```python
# AES of limits.
AES =[]
for limit in range(3):
    ask_size = df_trades['ask_size_' + str(limit + 1)]
    bid_size = df_trades["bid_size_" + str(limit + 1)]
    AES.append((ask_size.mean(), bid_size.mean() ))
print(AES)
```

```python
# We will use a scheme for the first limit, and then do the same for the others.
delta_bid_size = abs(df_trades.bid_size_1.diff(1).dropna())
delta_bid_price = abs(df_trades.bid_price_1.diff(1).dropna())
delta_index = pd.Series(np.array(np.sign(delta_bid_size + delta_bid_price)) *np.array(range(len(delta_bid_price))))
delta_index = (delta_index[delta_index > 0 ].astype(int))
events_queue_bid = df_trades.iloc[delta_index]
events_queue_bid.bid_size_1 = (events_queue_bid.bid_size_1 / AES[0][1]).astype(int) 
deltas = events_queue_bid.ts.diff(1).fillna(0)
events_queue_bid["deltas"] = deltas
events_queue_bid.bid_size_1.unique()

```

```python
cumulated_intensities = []
intensities_insertion = []
intensities_cancelation = []
intensities_marker_order = []
for n in range(10):
    events_to_use = events_queue_bid[events_queue_bid.bid_size_1 == n]
    cumulated_intensity = events_to_use.deltas.mean()
    cumulated_intensities.append(cumulated_intensity)
    length = len(events_to_use)

    if length > 0:
        intensity_insertion = cumulated_intensity * len(events_to_use[events_to_use.type == 1]) / length
        intensity_cancellation = cumulated_intensity * len(events_to_use[(events_to_use.type == 2) | (events_to_use.type == 3)]) / length
        intensity_market_order = cumulated_intensity * len(events_to_use[events_to_use.type == 4]) / length
        intensities_insertion.append(intensity_insertion)
        intensities_cancelation.append(intensity_cancellation)
        intensities_marker_order.append(intensity_market_order)    
x = range(n)
```

```python
# Now we will use it for the 3 best bids
bid_intensities_insertion = []
bid_intensities_cancelation = []
bid_intensities_market_order = []
for limit in range(1, 4):
    delta_bid_size = abs(df_trades["bid_size_" + str(limit)].diff(1).dropna())
    delta_bid_price = abs(df_trades["bid_price_" + str(limit)].diff(1).dropna())
    delta_index = pd.Series(np.array(np.sign(delta_bid_size + delta_bid_price)) *np.array(range(len(delta_bid_price))))
    delta_index = (delta_index[delta_index > 0 ].astype(int))
    events_queue_bid = df_trades.iloc[delta_index]
    events_queue_bid["bid_size_" + str(limit)] = (events_queue_bid["bid_size_" + str(limit)] / AES[limit - 1][1]).astype(int) 
    deltas = events_queue_bid.ts.diff(1).fillna(0)
    events_queue_bid["deltas"] = deltas
        
    cumulated_intensities = []
    intensities_insertion = []
    intensities_cancelation = []
    intensities_marker_order = []
    for n in range(12):
        events_to_use = events_queue_bid[(events_queue_bid["bid_size_" + str(limit)] / events_queue_bid["bid_size_" + str(limit)].mean()).astype(int) == n]
        cumulated_intensity = events_to_use.deltas.mean()
        cumulated_intensities.append(cumulated_intensity)
        length = len(events_to_use)
        if length >0:
            intensity_insertion = cumulated_intensity * len(events_to_use[events_to_use.type == 1]) / length
            intensity_cancellation = cumulated_intensity * len(events_to_use[(events_to_use.type == 2) | (events_to_use.type == 3)]) / length
            intensity_market_order = cumulated_intensity * len(events_to_use[events_to_use.type == 4]) / length
            intensities_insertion.append(intensity_insertion)
            intensities_cancelation.append(intensity_cancellation)
            intensities_marker_order.append(intensity_market_order)    
    bid_intensities_cancelation.append(intensities_cancelation)
    bid_intensities_insertion.append(intensities_insertion)
    bid_intensities_market_order.append(intensities_marker_order)


    
```

```python
# For ask: Please do not care for names

ask_intensities_insertion = []
ask_intensities_cancelation = []
ask_intensities_market_order = []

# Cas q_1 = 0
df_q_new = df_q[df_q.q_1 == 0].reset_index(drop=True)
new_AES = df_q_new.q2.mean()
delta_bid_size = abs(df_q_new.q2.diff(1).dropna())
delta_index = pd.Series(np.array(np.sign(delta_bid_size)) *np.array(range(len(delta_bid_size))))
delta_index = (delta_index[delta_index > 0 ].astype(int))
events_queue_bid = df_q_new.iloc[delta_index]
events_queue_bid["q2"] = (events_queue_bid["q2"] / new_AES).astype(int) 
deltas = events_queue_bid.ts.diff(1).fillna(0)
events_queue_bid["deltas"] = deltas
        
cumulated_intensities = []
intensities_insertion = []
intensities_cancelation = []
intensities_marker_order = []
for n in range(12):
    events_to_use = events_queue_bid[events_queue_bid["q2"] == n]
    cumulated_intensity = events_to_use.deltas.mean()
    cumulated_intensities.append(cumulated_intensity)
    length = len(events_to_use)
    if length > 1 :
        intensity_insertion = cumulated_intensity * len(events_to_use[events_to_use.type == 1]) / length
        intensity_cancellation = cumulated_intensity * len(events_to_use[(events_to_use.type == 2) | (events_to_use.type == 3)]) / length
        intensity_market_order = cumulated_intensity * len(events_to_use[events_to_use.type == 4]) / length
        intensities_insertion.append(intensity_insertion)
        intensities_cancelation.append(intensity_cancellation)
        intensities_marker_order.append(intensity_market_order)    
ask_intensities_cancelation.append(intensities_cancelation)
ask_intensities_insertion.append(intensities_insertion)
ask_intensities_market_order.append(intensities_marker_order)   

# Caq q_1 >0
df_q_new = df_q[df_q.q_1 > 0].reset_index(drop=True)
new_AES = df_q_new.q2.mean()

delta_bid_size = abs(df_q_new.q2.diff(1).dropna())
delta_index = pd.Series(np.array(np.sign(delta_bid_size)) *np.array(range(len(delta_bid_size))))
delta_index = (delta_index[delta_index > 0 ].astype(int))
events_queue_bid = df_q_new.iloc[delta_index]
events_queue_bid["q2"] = (events_queue_bid["q2"] / new_AES).astype(int) 
deltas = events_queue_bid.ts.diff(1).fillna(0)
events_queue_bid["deltas"] = deltas
        
cumulated_intensities = []
intensities_insertion = []
intensities_cancelation = []
intensities_marker_order = []
for n in range(12):
    events_to_use = events_queue_bid[(events_queue_bid["q2"] / events_queue_bid["q2"].mean()).astype(int) == n]
    cumulated_intensity = events_to_use.deltas.mean()
    cumulated_intensities.append(cumulated_intensity)
    length = len(events_to_use)
    if length > 1 :
        intensity_insertion = cumulated_intensity * len(events_to_use[events_to_use.type == 1]) / length
        intensity_cancellation = cumulated_intensity * len(events_to_use[(events_to_use.type == 2) | (events_to_use.type == 3)]) / length
        intensity_market_order = cumulated_intensity * len(events_to_use[events_to_use.type == 4]) / length
        intensities_insertion.append(intensity_insertion)
        intensities_cancelation.append(intensity_cancellation)
        intensities_marker_order.append(intensity_market_order)    
ask_intensities_cancelation.append(intensities_cancelation)
ask_intensities_insertion.append(intensities_insertion)
ask_intensities_market_order.append(intensities_marker_order)   
```

```python
ask_intensities_insertion = []
ask_intensities_cancelation = []
ask_intensities_market_order = []
count = 0
for k in range(0, len(limits) - 1):
    count+=1
    df_q.q_1 = (df_q.q_1 / (df_q.q_1.mean())).astype(int)
    df_q_new = df_q[(df_q.q_1 <= limits[k+1]) & (df_q.q_1 > limits[k])].reset_index(drop=True)
    
    new_AES = df_q_new.q2.mean()
    delta_bid_size = abs(df_q_new.q2.diff(1).dropna())
    delta_index = pd.Series(np.array(np.sign(delta_bid_size)) *np.array(range(len(delta_bid_size))))
    delta_index = (delta_index[delta_index > 0 ].astype(int))
    events_queue_bid = df_q_new.iloc[delta_index]
    events_queue_bid["q2"] = (events_queue_bid["q2"] / new_AES).astype(int) 
    deltas = events_queue_bid.ts.diff(1).fillna(0)
    events_queue_bid["deltas"] = deltas
        
    cumulated_intensities = []
    intensities_insertion = []
    intensities_cancelation = []
    intensities_marker_order = []
    for n in range(12):
        events_to_use = events_queue_bid[(events_queue_bid["q2"]/events_queue_bid["q2"].mean()).astype(int)   == n]
        cumulated_intensity = events_to_use.deltas.mean()
        cumulated_intensities.append(cumulated_intensity)
        length = len(events_to_use)
        if length > 1 :
            intensity_insertion = cumulated_intensity * len(events_to_use[events_to_use.type == 1]) / length
            intensity_cancellation = cumulated_intensity * len(events_to_use[(events_to_use.type == 2) | (events_to_use.type == 3)]) / length
            intensity_market_order = cumulated_intensity * len(events_to_use[events_to_use.type == 4]) / length
            intensities_insertion.append(intensity_insertion)
            intensities_cancelation.append(intensity_cancellation)
            intensities_marker_order.append(intensity_market_order)    
    ask_intensities_cancelation.append(intensities_cancelation)
    ask_intensities_insertion.append(intensities_insertion)
    ask_intensities_market_order.append(intensities_marker_order)  
    
    
```
